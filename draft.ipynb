```python
lfw_mask = r'lfw_mask_100'
dst_path= r'user_face\mask'
```

```python
import os
import shutil
```

```python
ls_user = os.listdir(lfw_mask)
count = 0
for user in ls_user:
    ls_img = os.listdir(os.path.join(lfw_mask, user))
    if len(ls_img) > 5:
        if not os.path.exists(os.path.join(dst_path, user)):
            os.mkdir(os.path.join(dst_path, user))
            
        for img in ls_img[:5]:
            src = os.path.join(os.path.join(lfw_mask, user), img)
            dst = os.path.join(os.path.join(dst_path, user), img)
            shutil.move(src, dst)

        count += 1
    
    if count == 100:
        break

    
```
